# Data Graph Explorer

```python
import matplotlib.pyplot as plt
import numpy
import pandas as pd
import random as r

#menu
print(".CSV Upload Options? \n",
            "1. Locally\n", 
            "2. Enter URL\n", 
            "3. Computer Selection\n", 
       "NOTE: The file must contain only numerical values in columns.")

while True:
    upload_data = int(input("How would you like to upload your data?"))
    if upload_data == 1:
        user_data = input("Enter the name (with .csv) or file path (with forward slashes [/]) of the file you wish to upload: ")
        break
    elif upload_data == 2:
        user_data = input("Enter the URL of the .csv file: ")
        break
    elif upload_data == 3:
        user_data = "https://calmcode.io/static/data/chickweight.csv"
        break
    else:
        print("Incorrect entry. Must enter a number from 1 to 3")


#save file as dataframe
data = pd.read_csv(user_data,  encoding = "ISO-8859-1")

#headings and first 2 rows
headers_2r = data.head(2)
print("\n", headers_2r)

#store column names as list
col_list = list(data.columns)

#/if only 2 data cols
if len(col_list) < 3: 
        x_name = r.choice(col_list)
        while True:
            y_name = r.choice(col_list)
            if x_name == y_name:
                continue
            else:
                break
        x = data[x_name].to_numpy()
        y = data[y_name].to_numpy()
        xmin = x.min() - 5
        xmax = x.max() + 5
        ymin = y.min() - 5
        ymax = y.max() + 5
        plt.axis([xmin,xmax,ymin,ymax]) 
        plt.plot(x, y, 'o')
        plt.xlabel(x_name)
        plt.ylabel(y_name)
        print(" ")
        plt.show()

else:
    #select random columns from list 
    x_name = r.choice(col_list)

    while True:
        y_name = r.choice(col_list)
        z_name = r.choice(col_list)
        if x_name == y_name:
            continue
        elif x_name==z_name or y_name==z_name:
            continue
        else:
            break

    #convert columns to numpy arrays  
    x = data[x_name].to_numpy()
    y = data[y_name].to_numpy()
    z = data[z_name].to_numpy()

    #1st plot
    xmin = x.min() - 2
    xmax = x.max() + 5
    ymin = y.min() - 2
    ymax = y.max() + 10
    fig, ax = plt.subplots()
    plt.axis([xmin,xmax,ymin,ymax])
    print(" ")
    plt.plot(x, y, 'go') 
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.show()

    #2nd plot
    xmin = x.min() - 2
    xmax = x.max() + 5
    zmin = z.min() - 2
    zmax = z.max() + 10

    fig, ax = plt.subplots()
    plt.axis([xmin,xmax,zmin,zmax])
    print(" ")
    plt.plot(x, z, 'o') 
    plt.xlabel(x_name)
    plt.ylabel(z_name)
    plt.show()


    #3rd plot
    ymin = y.min() - 2
    ymax = y.max() + 5
    zmin = z.min() - 2
    zmax = z.max() + 10

    fig, ax = plt.subplots()
    plt.axis([ymin,ymax,zmin,zmax])
    print(" ")
    plt.plot(y, z, 'ro')
    plt.xlabel(y_name)
    plt.ylabel(z_name)
    plt.show()
```
